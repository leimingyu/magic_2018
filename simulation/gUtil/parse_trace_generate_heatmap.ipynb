```python
# needed on mac
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def find_gUitl(file_csv, GpuNum=1, targetGpu=None):
    GpuNum = int(GpuNum)
    df_trace = pd.read_csv(file_csv,  engine='python')
    sampling_records = int(df_trace.shape[0] / GpuNum)
    
    util_array = None
    
    if targetGpu is None: # all the gpu trace
        util_array = np.zeros((GpuNum, sampling_records)) # each column is each time step
        #print('monitoring records array : {} x {}'.format(GpuNum, sampling_records))
    else:
        targetGpu = int(targetGpu) # make sure it is int
        util_array = np.zeros(sampling_records)
        #print util_array.shape
        util_array = util_array.reshape(-1, sampling_records)
        #print util_array.shape
    
    
    for index, row in df_trace.iterrows():
        #print index, row[0], row[1]
        gpu_util = float(row[1][:-2]) * 0.01 # read the utilization
    
        col_id =  int(index) / GpuNum
        row_id =  int(index) % GpuNum
    
        #print row_id, col_id, gpu_util
        if targetGpu is None:
            util_array[row_id, col_id] = gpu_util
        else:
            if row_id == targetGpu:
                util_array[0, col_id] = gpu_util
                
    return util_array
```

### traces is samples at 1s interval

```python
gpu_util = find_gUitl('fcfs_s1_max2.csv', GpuNum=12, targetGpu=6)
np.savetxt("gpuUtil_fcfs_max2.csv", gpu_util, delimiter=",")

sns.set()
# The default cmap is sns.cm.rocket. To reverse it set cmap to sns.cm.rocket_r
ax = sns.heatmap(gpu_util, yticklabels=False, cmap=sns.cm.rocket_r)
#fig = ax.get_figure()
#fig.savefig("test.png")
```
