### determine whether robust or sensitive to interference

```python
#
# generate a dataframe and save the results to csv to the disk
#

total_num = len(run2_app_speedup_dd.keys())

df = pd.DataFrame(index=np.arange(0, total_num), columns=['appName', 'Robust'])


TH = 0.8
rowID = 0
for app, app_speedup_dd in run2_app_speedup_dd.iteritems():
    print("working on app = {}".format(app))
    #print type(app_speedup_dd)
    app_speedup_array = np.array([run2[0] for run2 in app_speedup_dd.values()])# get the 1st speedup (for target app)
    #print app_speedup_array
    #print app_speedup_array.min(), app_speedup_array.max()
    #print app_speedup_array.mean(), app_speedup_array.std()
    
    print("avg speedup = {}".format(app_speedup_array.mean()))
    robust = 1
    if app_speedup_array.mean() >= TH:
        print "robust\n"
    else:
        print "sensitive\n"
        robust = 0
        
    #
    # update dataframe
    #
    df.loc[rowID, 'appName'] = app
    df.loc[rowID, 'Robust']  = robust
    rowID += 1
    #break
```
